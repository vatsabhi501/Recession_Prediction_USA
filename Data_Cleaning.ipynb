```python
yield_df['T10Y3M'].replace(".","0", inplace= True)
```

```python
# converting DATE column to string and value column into float
yield_df["DATE"] = yield_df["DATE"].astype("str")
yield_df = yield_df.astype({'T10Y3M':'float'})
```

```python
# adding Month and Year column
yield_df["Month"] = yield_df["DATE"].apply(lambda x: x.split("-")[1])
yield_df["Year"] = yield_df["DATE"].apply(lambda x: x.split("-")[0])
```

```python
yield_df
```

```python
#Initialising the Monthly Average Columns to 0 and retaining the NaN values
yield_df['Avg T10Y3M'] = yield_df['T10Y3M'].apply(lambda x: float("NaN") if math.isnan(x) else 0)
```

```python
count = 0
sum_ = 0

for row in range(len(yield_df.index)-1):

  #Monthly averages for T10Y3M column
  if math.isnan(yield_df['T10Y3M'][row]) == False:
      count += 1.0
      sum_ += yield_df['T10Y3M'][row]
      avg_T10Y3M = float(sum_ / count) #this is storing the current average
      
      if yield_df['Month'][row]!=yield_df['Month'][(row+1)]:
        yield_df['Avg T10Y3M'][row] = avg_T10Y3M
        sum_ = 0.0 #resetting the sums and averages to 0 to iterate for next month
        avg_T10Y3M = 0.0
        count = 0.0
        
```

```python
yield_df
```

```python
yield_df = yield_df.loc[~((yield_df['Avg T10Y3M'] == 0))]
```

```python
yield_df
```

```python
# drop T10Y3M column from yield_df dataset
yield_df.drop(["DATE"], axis=1, inplace= True)
yield_df.drop(["T10Y3M"], axis=1, inplace= True)
```

# Merging the datasets


```python
# converting "Month" and "Year" columns as "int"
inflation_cpi_df['Year'] = inflation_cpi_df['Year'].astype(int)
inflation_cpi_df['Month']=inflation_cpi_df['Month'].astype(int)
unrate_df['Year']=unrate_df['Year'].astype(int)
unrate_df['Month']=unrate_df['Month'].astype(int)
cci_df['Year']=cci_df['Year'].astype(int)
cci_df['Month']=cci_df['Month'].astype(int)
fedfunds_df['Year']=fedfunds_df['Year'].astype(int)
fedfunds_df['Month']=fedfunds_df['Month'].astype(int)
buffett_df['Year']=buffett_df['Year'].astype(int)
buffett_df['Month']=buffett_df['Month'].astype(int)
m2_velocity['Month']=m2_velocity["Month"].astype(int)
m2_velocity['Year']=m2_velocity["Year"].astype(int)
manufacturing_output['Month']=manufacturing_output['Month'].astype(int)
manufacturing_output['Year']=manufacturing_output['Year'].astype(int)
debt_to_gdp['Month']=debt_to_gdp['Month'].astype(int)
debt_to_gdp['Year']=debt_to_gdp['Year'].astype(int)
housing_price['Month']=housing_price['Month'].astype(int)
housing_price['Year']=housing_price['Year'].astype(int)
yield_df['Month']=yield_df['Month'].astype(int)
yield_df['Year']=yield_df['Year'].astype(int)
```

```python
data_frames = [inflation_cpi_df,unrate_df,fedfunds_df, cci_df,buffett_df,yield_df,m2_velocity,manufacturing_output,debt_to_gdp,housing_price]
```

```python
data_frames
```

```python
df = reduce(lambda left,right: pd.merge(left,right,on=['Month','Year'], how='outer'), data_frames)
```

#### According to Forbes (https://www.forbes.com/sites/billconerly/2022/05/12/the-yield-curve-predicts-recession-sometimes-but-not-today/?sh=7dd995365443), "We can measure the yield curve by comparing any two maturities, but the most common measure compares the 10-year Treasury bond to the 3-month Treasury bill. "
Thus I will be removing all other yield curve data except the data that corresponds to the 10-year Treasury bond and the 3-month Treasury bill.

```python
df
```

```python
#Handling Buffet Indicator, which has values after 3 month gaps. Giving those '3 month gaps' the value of the month prior to that gap.
for i in range(288, len(df)): #because all indices less than 288 are NaN values for the Buffett Indicator. 
  if math.isnan(df['Market_Cap_to_GDP'][i]):
    df['Market_Cap_to_GDP'][i] = df['Market_Cap_to_GDP'][i-1]
  if math.isnan(df['M2V'][i]):
    df['M2V'][i] = df['M2V'][i-1]
  if math.isnan(df['Manufacturing Output'][i]):
    df['Manufacturing Output'][i] = df['Manufacturing Output'][i-1]
  if math.isnan(df['Debt_to_GDP'][i]):
    df['Debt_to_GDP'][i] = df['Debt_to_GDP'][i-1]
  if math.isnan(df['Housing Price'][i]):
    df['Housing Price'][i] = df['Housing Price'][i-1]
    
```
